```python
#using pandas applying oneHot Encoding
import pandas as pd
data={'Car':['Tyota','Carola','Honda','Civic']}
df=pd.DataFrame(data)
df
```

```python
df_encoded=pd.get_dummies(df,dtype='int')
df_encoded
```

```python
from sklearn .preprocessing import OneHotEncoder
import numpy as np
enco=OneHotEncoder(handle_unknown='ignore')                              #(creating the encoder)
X=[['Tyota'],['Carola'],['Honda'],['Civic']]      #(apny data ko verible asigned kiya hai , sample data)
enco.fit(X)                                                             #fit the encoder to data         
Result=enco.transform([['Tyota']]).toarray()               #transforming new data to create array
Result
```
